```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('titanic_toy.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train
```

```python
x_train['Age_99']=x_train['Age'].fillna(99)
x_train['Age_minus1']=x_train['Age'].fillna(-1)
x_train['Fare_999']=x_train['Fare'].fillna(999)
x_train['Fare_minus1']=x_train['Fare'].fillna(-1)
```

```python
print(f" original age varience ={x_train['Age'].var()}")
print(f" after impute 99 age varience ={x_train['Age_99'].var()}")
print(f" after impute -1  age varience ={x_train['Age_minus1'].var()}")

print(f" original fare varience ={x_train['Age'].var()}")
print(f" after impute 999 fare varience= {x_train['Fare_999'].var()}")
print(f" after impute -1  fare varience ={x_train['Fare_minus1'].var()}")
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

x_train['Age'].plot(kind='kde',ax=ax)
x_train['Age_99'].plot(kind='kde',color='red',ax=ax)
x_train['Age_minus1'].plot(kind='kde',color='green',ax=ax)
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

x_train['Fare'].plot(kind='kde',ax=ax)
x_train['Fare_999'].plot(kind='kde',color='red',ax=ax)
x_train['Fare_minus1'].plot(kind='kde',color='green',ax=ax)
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

# arbitiary impute using sklearn


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
trf=ColumnTransformer([
    ('impute1',SimpleImputer(strategy='constant',fill_value=99),['Age']),
        ('impute2',SimpleImputer(strategy='constant',fill_value=999),['Fare'])

],remainder='passthrough')
```

```python
trf.fit(x_train)
```

```python
trf.named_transformers_['impute1'].statistics_
```

```python
trf.named_transformers_['impute2'].statistics_
x_train
```

```python
x_train=trf.transform(x_train)
x_test=trf.transform(x_test)
```
